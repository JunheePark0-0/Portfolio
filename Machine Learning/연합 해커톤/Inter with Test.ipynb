```python
from pycaret.classification import setup
from pycaret.classification import *
```

```python
train_filtered = pd.read_csv('/content/drive/MyDrive/DATA/train_filtered.csv')
```

```python
# click이 0인 데이터 중 50%를 샘플링
df = train_filtered.copy()
click_0_sample = df[df['click'] == 0].sample(frac=0.01, random_state=42)

# click이 1인 데이터 중 50%를 샘플링
click_1_sample = df[df['click'] == 1].sample(frac=0.01, random_state=42)

# 두 샘플을 합쳐서 축소된 데이터셋 생성
reduced_df = pd.concat([click_0_sample, click_1_sample]).reset_index(drop=True)
reduced_df.to_csv("/content/drive/MyDrive/DATA/reduced_inter.csv", index=False)
```
